```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame and make the index the "coin_id" column.
market_data_df = pd.read_csv("Resources/crypto_market_data.csv", index_col="coin_id")

# Display sample data
market_data_df.head(10)
```

```python
market_data_df.info()
len(market_data_df)
```

```python
# Generate summary statistics
market_data_df.describe()
```

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
numeric_columns = market_data_df.columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(market_data_df[numeric_columns])
scaled_df = pd.DataFrame(scaled_data, columns=numeric_columns, index=market_data_df.index)
scaled_df.head()
```

```python
# Create a DataFrame with the scaled data
# Already completed above

# Copy the crypto names from the original data
# already completed above


# Set the coinid column as index
# Already done

# Display sample data
scaled_df.head()
```

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_values = list(range(1, 11))

# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values:
    
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, n_init='auto', random_state=3)
    
# 2. Fit the model to the data using the scaled DataFrame
    model.fit(scaled_df)
    
# 3. Append the model.inertia_ to the inertia list
    inertia_values.append(model.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,
    'inertia': inertia_values
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)

# Display the DataFrame
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(10, 6))
plt.plot(elbow_df['k'], elbow_df['inertia'], 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.grid(True)
plt.show()
```

#### Answer the following question: 
**Question:** What is the best value for `k`?

**Answer:**    The best answer is 5 which is the last movement in the chart that shows a substantial decrease.   After 5 the decrease in inertia is very little and with too many groups it won't tell us very much more information

### Cluster Cryptocurrencies with K-means Using the Original Scaled Data.

```python
# Initialize the K-Means model using the best value for k
best_k_value = 5
```

```python
# Fit the K-Means model using the scaled data
kmeans_model = KMeans(n_clusters = best_k_value, n_init='auto', random_state=3)
kmeans_model.fit(scaled_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
cluster_predictions = kmeans_model.predict(scaled_df)

# View the resulting array of cluster values.
cluster_predictions
```

```python
# Create a copy of the DataFrame
scaled_with_cluster = scaled_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
scaled_with_cluster['Cluster'] = cluster_predictions

# Display sample data
scaled_with_cluster
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` on the original scaled DataFrame to reduce to three principal components.
pca_data = pca_model.fit_transform(scaled_df)

# View the first five rows of the DataFrame. 
print("array = ", pca_data[:5])
```

```python
# Retrieve the explained variance to determine how much information  can be attributed to each principal component.
explained_variance = pca_model.explained_variance_ratio_

print("Explained variance ratio:", explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The total would be 37.19% + 34.70% + 17.60% or in other words, 89.49% of the original variance has been retained when we turned this into a three dimensional space of the principal components that PCA created.

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(
    data=pca_data,
    columns=['PC1', 'PC2', 'PC3']
)

# Copy the crypto names from the original data
pca_df['coin_id'] = scaled_df.index



# Set the coinid column as index

pca_df.set_index('coin_id', inplace=True)


# Display sample data
pca_df.head(10)
```

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values to try
# Use a range from 1 to 11
k_values_for_pca = list(range(1, 11))

# Create an empty list to store the inertia values
inertia_values_for_pca = []

# Create a for loop to compute the inertia with each possible value of k
for k in k_values_for_pca:

# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
    model_for_pca = KMeans(n_clusters = k, n_init="auto", random_state=3) 

# 2. Fit the model to the data using PCA DataFrame.
    model_for_pca.fit(pca_df)

# 3. Append the model.inertia_ to the inertia list
    inertia_values_for_pca.append(model_for_pca.inertia_)

# Create a dictionary with the data to plot the Elbow curve
elbow_data_for_pca = {
    'k': k_values_for_pca,
    'inertia': inertia_values_for_pca
}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df_for_pca = pd.DataFrame(elbow_data_for_pca)

# Display the DataFrame
elbow_df_for_pca
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
plt.figure(figsize=(10, 6))
plt.plot(elbow_df_for_pca['k'], elbow_df_for_pca['inertia'], 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.grid(True)
plt.show()
```

#### Answer the following questions: 
* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**    Four - this is plain to see and even clearer than the earlier elbow curve.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:**   Yes.  It is four, indicating that when breaking it down to principal components (keeping most of the variance) there is a more natural division into 4 clusters rather than 5
